```python

import pandas as pd # library for data analsysis
import numpy as np # library to handle data in a vectorized manner
import random # library for random number generation

# libraries for displaying images
from IPython.display import Image 
from IPython.core.display import HTML 

!conda install -c conda-forge folium=0.5.0 --yes
import folium # plotting library

print('Folium installed')
print('Libraries imported.')
```

```python
# read html table to dataframe

dfs = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M', header=0)
dfs[0].head()
```

```python
#drop rows where Borough is Not assigned
dfm=dfs[0]
dfm.head()

dfm.drop(dfm[dfm['Borough'] == 'Not assigned'].index, inplace = True)

# reset index, because we droped rows
dfm.reset_index(drop=True, inplace=True)

dfm.set_index(["Postal Code"], inplace = True, 
                    append = True, drop = True) 

dfm.head()
```

```python
#look at shape
dfm.shape
```

```python
#reading csv file with long/lat info
dfss = pd.read_csv('http://cocl.us/Geospatial_data')


dfss.set_index(["Postal Code"], inplace = True, 
                    append = True, drop = True) 
dfss.head()
```

```python
#Join the two dataframes

merged_df = pd.merge(dfm, dfss, 
                     left_on = 'Postal Code', 
                     right_on = 'Postal Code', 
                     how='left')
merged_df
```

```python
# create map centered around Downtown Toronto
```

```python
map = folium.Map(location=[43.6, -79.3], zoom_start=11)

# add markers to map

for lat, lng, label in zip(merged_df['Latitude'], merged_df['Longitude'], merged_df['Borough'], merged_df ['Neighborhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map)  
    
map
```
